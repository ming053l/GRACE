```python
# FF++
type_list = ['Face2Face', 'NeuralTextures','Deepfakes', 'FaceSwap']
# type_list = ['Deepfakes', 'FaceSwap']
for tp in type_list:
    pth='/hdd4/FF_src/c40/manipulated_sequences/{}/c40/videos/'.format(tp)
    # pth = '/hdd4/FF_src/raw/original_sequences/actors/raw/videos/'
    dx = glob.glob(pth+'*.mp4')
    train_data, test = train_test_split(dx, random_state=777, train_size=0.8)
    val_data, test_data = train_test_split(test, random_state=777, train_size=0.5)
    data_name = ['train', 'val', 'test']
    data_use = [train_data, val_data, test_data]
    for num, dx in enumerate(data_use):
        outdir_base = '/hdd1/DeepFakes_may/FF++/c40'
        outdir_use = data_name[num]+'/'+tp
        outdir = os.path.join(outdir_base,outdir_use )
        for filename in dx: 
            v_cap = cv2.VideoCapture(filename)
            v_len = int(v_cap.get(cv2.CAP_PROP_FRAME_COUNT))
            # (major_ver, minor_ver, subminor_ver) = (cv2.__version__).split('.')
            print(filename,v_len)
            frame_interval = 1
            path = '/hdd1/DeepFakes_may/FF++/c40/{}.txt'.format(data_name[num])
            half_len=v_len//2

            fn = os.path.basename(filename)[:-4]
            count = 0
            if not os.path.exists(outdir+'/'+fn):
                os.makedirs(outdir+'/'+fn)

            with open(path, 'a') as f:
                f.write(os.path.join(outdir_use, fn)+' 1'+'\n')
                
            path_ =  os.path.join(outdir, fn, 'list.txt')

            do_save = 0
            if v_len<=100:
                cut_num = v_len
                save_num = 0
            else:
                cut_num = half_len+50
                save_num = half_len-50

            for j in range(cut_num):
                # Load frame
                success, frame = v_cap.read()
                do_save += 1
                if do_save>=(save_num):
                    if not success:
                        continue
                    if j % frame_interval==0:
                        cv2.imwrite(os.path.join(outdir, fn, '%03d.png' % (count)),frame)
                        with open(path_, 'a') as f:
                            f.write('%03d.png' % (count)+'\n')
                        count+=1
                        # break
    f.close()
```
